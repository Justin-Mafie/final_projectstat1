```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Load the dataset
df = pd.read_csv("owid-covid-data.csv")

# Assuming 'total_deaths' is the only feature and 'deaths' is the target variable
X = df[['total_deaths']]
y = df['total_deaths']

# Handle missing values
X = X.fillna(0)  # Replace NaN values with 0 (or use another strategy based on your data)
y = y.fillna(0)

# Define a threshold for classifying deaths
threshold = 100

# Convert to a binary classification task
y_class = (y > threshold).astype(int)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_class, test_size=0.2, random_state=42)

# Standardize/Normalize data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Build the neural network model
model = Sequential([
    Dense(64, activation='relu', input_dim=X_train.shape[1]),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')  # Use sigmoid activation for binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model
loss, accuracy, precision, recall = model.evaluate(X_test, y_test)

# Handle NaN values in metrics
if np.isnan(accuracy) or np.isnan(precision) or np.isnan(recall):
    print("Error: Metrics contain NaN values. Check your data and model architecture.")
else:
    print(f"Accuracy: {accuracy}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Load the dataset
df = pd.read_csv("owid-covid-data.csv")

# Assuming 'total_deaths' is the only feature and 'deaths' is the target variable
X = df[['total_deaths']]
y = df['total_deaths']

# Handle missing values
X = X.fillna(0)  # Replace NaN values with 0 (or use another strategy based on your data)
y = y.fillna(0)

# Define a threshold for classifying deaths
threshold = 100

# Convert to a binary classification task
y_class = (y > threshold).astype(int)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_class, test_size=0.2, random_state=42)

# Standardize/Normalize data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Build the neural network model
model = Sequential([
    Dense(64, activation='relu', input_dim=X_train.shape[1]),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')  # Use sigmoid activation for binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model
loss, accuracy, precision, recall = model.evaluate(X_test, y_test)

# Calculate and print F1 score
y_pred = (model.predict(X_test) > 0.5).astype(int)
f1 = f1_score(y_test, y_pred)

# Handle NaN values in metrics
if np.isnan(accuracy) or np.isnan(precision) or np.isnan(recall) or np.isnan(f1):
    print("Error: Metrics contain NaN values. Check your data and model architecture.")
else:
    print(f"Accuracy: {accuracy}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f1}")
```
